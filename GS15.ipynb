```python

# Faut pas prendre en compte les fautes mon clavier n'a pas les accent, bien Merci !!

# En cour de codage
def chiffrement_dechiffrement(Message, Cle): # COBRA
    # ADD round 1
    data = ''.join(str(int(Message[i]) ^ int(Cle[i % len(Cle)])) for i in range (len(Message)))  # Xor avec la cle d'iteration reste a voir ce que c'est  
    return data #Faudra voir comment decouper data en bloc de 32
# il pourrait y avoir une bloque for de 32 iteration pour les 32 round de permutation SBOX


# Exemple d'utilisation
#data = "11010011100101101010011111000011"  # Exemple de 32 bits en binaire
#round_num = 10  # Exemple de numéro de ronde
#result = apply_sboxes_parallel(data, round_num)
#print("Données après substitution avec S-box:", result)






    # SBOX 
    # Quatre S-boxes de 4 bits (16 valeurs possibles pour chaque S-box)

#Générer une clé de N bits // ok mais a modifier
import random as rd
def Key_of_N_bits(N): # Générer une clé de N bits
  return ''.join(str(rd.randint(0,1)) for i in range (N))

Key = Key_of_N_bits(128)
Key


#Conversion du fichier de caractères en fichier de binaires: // ok

def Text_To_Binary(Fichier_Text, Nom_fichier):
  with open(Fichier_Text, 'r', encoding='utf-8') as fichier:
   Message = fichier.read()
  Message_bin = ''.join(format(ord(caractere), '08b') for caractere in Message)
  with open('fichier_binaire.txt', 'w', encoding='utf-8') as fichier_binaire:
    fichier_binaire.write(Message_bin)
  return Message_bin




# Creation de SBOX  // en cour de codage
S_BOXES = [
    [0x6, 0x4, 0xC, 0x5, 0x0, 0x7, 0x2, 0xE, 0x1, 0xF, 0x3, 0xD, 0xA, 0x9, 0xB, 0x8],  # S-box 1
    [0xF, 0xC, 0x2, 0x8, 0x4, 0x9, 0x1, 0xA, 0x0, 0xE, 0x6, 0x7, 0xB, 0x5, 0x3, 0xD],  # S-box 2
    [0x8, 0x6, 0x7, 0x9, 0x3, 0xC, 0xA, 0x2, 0x5, 0xF, 0x4, 0x0, 0xD, 0xE, 0x1, 0xB],  # S-box 3
    [0xA, 0x4, 0x1, 0xD, 0x0, 0x5, 0xC, 0x7, 0x6, 0xF, 0x8, 0xE, 0xB, 0x9, 0x3, 0x2]   # S-box 4
]

def apply_sbox(value, sbox_index):
    """Applique la S-box spécifiée sur une valeur de 4 bits."""
    sbox = S_BOXES[sbox_index]
    return sbox[value]

def apply_sboxes_parallel(data, round_num):
    """
    Applique les S-boxes en parallèle sur 32 blocs de 4 bits dans 'data'.
    Choisit la S-box en fonction du numéro de ronde.
    """
    # Sélectionner la S-box en fonction du numéro de la ronde
    sbox_index = round_num // 8  # 0 pour 0-7, 1 pour 8-15, 2 pour 16-23, 3 pour 24-31
    sbox = S_BOXES[sbox_index]
    
    # Appliquer la S-box sur chaque bloc de 4 bits
    transformed_data = []
    for i in range(0, len(data), 4):  # Traite chaque bloc de 4 bits
        block = int(data[i:i+4], 2)  # Convertir le bloc en entier (0-15)
        transformed_block = sbox[block]  # Appliquer la S-box
        transformed_data.append(f'{transformed_block:04b}')  # Convertir en binaire sur 4 bits
    
    return ''.join(transformed_data)









#Affichage d'un message binaire en message avec caractères // ok

def Bin_To_Text(Texte_binaire, Nom_fichier):
  Message_Code = ' '.join(Texte_binaire[i : i+8] for i in range(0, len(Texte_binaire), 8))
  Octets = Message_Code.split()
  Message_decode = ''.join(chr(int(groupe, 2)) for groupe in Octets)
  #On écrit le message décodé dans un fichier (Facultatif)
  with open(Nom_fichier, 'w', encoding='utf-8') as fichier_message:
    fichier_message.write(Message_decode)
  print(Message_decode)

#Bin_To_Text(Message_chiffre, "Faux_Message.txt")



# en cour  de codage 
def interface():  #interface pour montrer les fonctions de façon unitaire 
    nom = input(" Entrer votre nom:")
    print("Bonjour ",nom ,"! Que souhaitez vous faire aujourd'hui ?\n\n"+
      "-->1<-- Chiffrer/déchiffrer des messages. \n"+
      "-->2<-- Créer un couple de clés publique / privée (générer un grand nombre premier) \n"+
      "-->3<-- Hashage \n"
      "-->4<-- Preuve de travail \n"+
      "-->5<-- Génération et la vérification d’une signature \n"+ 
      "-->6<-- Création d’une blockchain \n"+
      "-->7<-- incrémentation et vérification d’une blockchain")
     #switcher entre les cas 
    cas = int(input())

    if cas == 1:
         print( chiffrement_dechiffrement(Text_To_Binary('C:/Users/USER/Desktop/poeme.txt' , 'Fichier_Binaire.txt'), Key) )
    elif cas == 2:
       print(" A completer")
    else:
      print("Cas non trouvé")
    
interface()   





     














```
